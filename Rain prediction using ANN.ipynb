```python
#Name: Arundarasi Rajendran
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("weatherAUS.csv")
```

```python
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
df1 = data.drop(axis=1,columns=['Evaporation', 'Sunshine','Cloud9am','Cloud3pm'])
```

```python
df1
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df1.info()
```

```python
df1['Month'] = df1['Date'].dt.month
df1['Year'] = df1['Date'].dt.year
df1['Day'] = df1['Date'].dt.day
```

```python
df1
```

```python
df2 = df1.dropna(how='any',axis=0)
```

```python
df2
```

```python
df2.describe()
```

```python
df2["RainToday"] = df2["RainToday"].map({"No":0,"Yes":1})
df2["RainTomorrow"] = df2["RainTomorrow"].map({"No":0,"Yes":1})
df2.head()
```

```python
WindGustDir_dummies = pd.get_dummies(df2["WindGustDir"],drop_first=True,prefix="WindGustDir")
WindDir9am_dummies = pd.get_dummies(df2["WindDir9am"],drop_first=True,prefix="WindDir9am")
WindDir3pm_dummies = pd.get_dummies(df2["WindDir3pm"],drop_first=True,prefix="WindDir3pm")
Location_dummies = pd.get_dummies(df2["Location"],drop_first=True,prefix="Country")
df3=pd.concat([df2,Location_dummies, WindDir3pm_dummies,WindDir9am_dummies,WindGustDir_dummies],axis=1)
df3.head()
```

```python
df3.drop(["Location","Date", "WindGustDir","WindDir9am","WindDir3pm"],inplace=True,axis=1)
```

```python
df3
```

```python
cols = list(df3.columns.values) #Make a list of all of the columns in the df
cols.pop(cols.index('RainTomorrow')) #Remove b from list
df3 = df3[cols+["RainTomorrow"]]
df3.head()
```

```python
X = df3.iloc[:, :-1].values
y = df3.loc[:,"RainTomorrow"].values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
#dense
#add layer to layer into our model
```

```python
model = Sequential()
```

```python
model.add(Dense(input_dim=104,kernel_initializer='random_uniform',units=15,activation='relu'))
```

```python
model.add(Dense(units=1,kernel_initializer='random_uniform',activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=20,batch_size=32)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred=(y_pred>=0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
